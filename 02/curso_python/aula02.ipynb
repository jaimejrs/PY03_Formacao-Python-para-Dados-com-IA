```python
entrada = input('Insira uma lista de números separados por espaço: ')
numeros = entrada.split() 

pares = 0
impar = 0

for item in numeros:
    number = int(item)  
    if number % 2 == 0:
        pares += 1 
        print(number, 'é par')
    else:
        impar += 1
        print(number, 'é ímpar')
print(f'existem {pares} numeros pares e {impar} numeros impar')
```

```python
carrinho = []

print("LISTA PRODUTOS")

while True:
    produto = input('Nome do produto | DIGITE"sair" para encerrar): ')
    if produto.lower() == 'sair': 
        break

    quantidade = int(input(f'Quantidade de {produto}: '))
    preco_unitario = float(input(f'Preço unitário de {produto}: R$ '))
    
    total_item = quantidade * preco_unitario
    
    # Adiciona uma tupla (pacote de dados) na lista do carrinho
    carrinho.append((produto, quantidade, preco_unitario, total_item))

print("RELATÓRIO DE COMPRAS")
print(f"{'PRODUTO':<20} | {'QTD':^5} | {'UNITÁRIO':^10} | {'TOTAL':^10}")

valor_final_compra = 0

for item in carrinho:
    nome = item[0]
    qtd = item[1]
    unit = item[2]
    tot = item[3]
    
    valor_final_compra += tot
    
    # .2f formata para 2 casas decimais
    print(f"{nome:<20} | {qtd:^5} | R$ {unit:>7.2f} | R$ {tot:>7.2f}")

print(f"{'TOTAL A PAGAR:':<41} R$ {valor_final_compra:.2f}")
```

```python
carrinho = [] 

print("LISTA DE PRODUTOS")

while True:
    produto = input('Nome do produto - "SAIR" para encerrar: ')
    
    if produto.lower() == 'sair':
        break

    quantidade = int(input(f'Quantidade de {produto}: '))
    preco_unitario = float(input(f'Preço unitário de {produto}: R$ '))
    
    total_item = quantidade * preco_unitario
    
    item_atual = [produto, quantidade, preco_unitario, total_item]
    carrinho.append(item_atual)
    

print(f"{'PRODUTO'} | {'QTD'} | {' VALOR UNITÁRIO'} | {'TOTAL'}")


valor_final_compra = 0

for item in carrinho:
    nome = item[0]
    qtd = item[1]
    unit = item[2]  
    tot = item[3]
    
    valor_final_compra += tot
    
    print(f"{nome} | {qtd} | R$ {unit} | R$ {tot}")

print(f"{'TOTAL:'} R$ {valor_final_compra}")
```
