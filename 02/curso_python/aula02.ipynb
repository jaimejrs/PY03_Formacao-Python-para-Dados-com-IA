```python
entrada = input('Insira uma lista de números separados por espaço: ')
numeros = entrada.split() 

pares = 0
impar = 0

for item in numeros:
    number = int(item)  
    if number % 2 == 0:
        pares += 1 
        print(number, 'é par')
    else:
        impar += 1
        print(number, 'é ímpar')
print(f'existem {pares} numeros pares e {impar} numeros impar')
```

```python
entrada = input('Insira uma lista de números separados por espaço: ')
numeros = entrada.split() 

pares = 0
impar = 0

for item in numeros:
    number = int(item)  
    if number % 2 == 0:
        pares += 1      # CORRIGIDO: O sinal de + vem antes do =
        print(number, 'é par')
    else:
        impar += 1      # CORRIGIDO: O sinal de + vem antes do =
        print(number, 'é ímpar')

print(f'Existem {pares} números pares e {impar} números ímpares')
```
